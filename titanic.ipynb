```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
from tensorflow import keras
```

```python
train = pd.read_csv(r'C:\Users\ariha\OneDrive - IIT Kanpur\Desktop\train.csv')
train.head()
```

```python
sex_binary_test = pd.read_csv(r'C:\Users\ariha\OneDrive - IIT Kanpur\Desktop\gender_submission.csv')
sex_binary_test.head()
```

```python
test = pd.read_csv(r'C:\Users\ariha\OneDrive - IIT Kanpur\Desktop\test.csv')
test.head()
```

```python
x_train = train[['Pclass','Sex','Age','Fare','Embarked','SibSp','Parch']]
x_train['Sex'] = x_train['Sex'].replace({'male':1,'female':0})
x_train['Embarked'].replace('Q', 0,inplace=True)
x_train['Embarked'].replace('S', 1,inplace=True)
x_train['Embarked'].replace('C', 2,inplace=True)
x_train['Age'] = x_train['Age']/100
x_train.head()
```

```python
y_train = train['Survived']
y_train = y_train.to_numpy()
y_train
```

```python
x_test = test[['Pclass','Sex','Age','Fare','Embarked','SibSp','Parch']]
x_test = x_test.replace({'male':1,'female':0})
x_test['Embarked'].replace('Q', 0,inplace=True)
x_test['Embarked'].replace('S', 1,inplace=True)
x_test['Embarked'].replace('C', 2,inplace=True)
x_test['Age'] = x_test['Age']/100
x_test.head()
```

```python
x_train = x_train.fillna(x_train.mean())
len(x_train)
```

```python
x_test = x_test.fillna(x_test.mean())
len(x_test)
```

```python
x_train_arr = x_train.to_numpy()
x_test_arr = x_test.to_numpy()
```

```python
x_train_arr
```

```python
x_test_arr
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(7,), activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train_arr, y_train, epochs=1000)
```

```python
model.evaluate(x_train_arr,y_train)
```

```python
predict = model.predict(x_train_arr)
predict = predict>0.5
predict = 1*predict
predict = predict.reshape(-1,)
predict
```

```python
predict = model.predict(x_test_arr)
predict = predict>0.5
predict = 1*predict
predict = predict.reshape(-1,)
Id = test.PassengerId
Id = pd.DataFrame(Id)
predict = pd.DataFrame(predict)
predict.columns = ['Survived']
Id,predict
```
