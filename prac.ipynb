```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv(r'C:\Users\ariha\OneDrive - IIT Kanpur\Desktop\train.csv')
df.head()
```

```python
input = df[['Pclass','Sex','Age','Fare']]
target = df['Survived']
```

```python
input.head()
```

```python
target.head()
```

```python
le_sex = LabelEncoder()
```

```python
input['sex_bin'] = le_sex.fit_transform(input['Sex'])
input=input.drop('Sex',axis='columns')
```

```python
from sklearn import tree
from sklearn.metrics import confusion_matrix
model = tree.DecisionTreeClassifier(criterion='entropy')
model.fit(input,target)
model.score(input,target)
```

```python
test = pd.read_csv(r'C:\Users\ariha\OneDrive - IIT Kanpur\Desktop\test.csv')
test
```

```python
test_input = test[['Pclass','Sex','Age','Fare']]
test_input['sex_bin'] = le_sex.fit_transform(test_input['Sex'])
test_input=test_input.drop('Sex',axis='columns')
test_input
```

```python
prediction=model.predict(test_input)
prediction
```

```python
prediction.tofile('submitfirst.csv', sep = ',')
```

```python
output = pd.DataFrame(test.PassengerId)
output.insert(1,'Survived',prediction)
output.head()
```
